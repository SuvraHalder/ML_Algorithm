```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```

```python
#Lets Apply Gradient Decent Asuuming Slope is constant  m = 78.35
# and lets assume the starting value for intercept  b =0
b=0
y_pred = ((78.35 * x)+b).reshape(4)
plt.plot(x,lr.predict(x),color='red',label='Ols')
plt.scatter(x,y)
plt.plot(x,y_pred,color='blue',label='b=0(GD)')
plt.legend()
plt.show()

'''
  We can see For b = 0 , it is not god is best fit  line .
  It means  we need to optimize it .

'''
```

```python
# Apply GD B_new  = B_old - (slope * learning rate)

# Iteration 1

Slope = - 2 * np.sum(y - 78.35 * x - b)
learning_rate =  0.01
b = b - (learning_rate * Slope)

y_pred = ((78.35 * x)+b).reshape(4)
plt.plot(x,lr.predict(x),color='red',label='Ols')
plt.scatter(x,y)
plt.plot(x,y_pred,color='blue',label='b=0(GD)')
plt.legend()
plt.show()

'''
We Can  see it got much closer . Compared to Random Asuumed Value .

'''
```
